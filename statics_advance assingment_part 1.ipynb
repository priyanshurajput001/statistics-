```python
# 2. Discrete Uniform Distribution & PMF
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

low, high = 1, 7  # dice roll 1 to 6
x = np.arange(low, high)
pmf = randint.pmf(x, low, high)

plt.bar(x, pmf)
plt.title("Discrete Uniform Distribution (PMF)")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

```python
# 3. Bernoulli Distribution PDF
from scipy.stats import bernoulli

def bernoulli_pdf(p):
    x = [0, 1]
    pdf = bernoulli.pmf(x, p)
    for val, prob in zip(x, pdf):
        print(f"P(X={val}) = {prob}")

bernoulli_pdf(0.3)
```

```python
# 4. Binomial Distribution (n=10, p=0.5) Histogram
from scipy.stats import binom

n, p = 10, 0.5
data = binom.rvs(n, p, size=1000)

plt.hist(data, bins=np.arange(0, n+2)-0.5, edgecolor='black', density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5. Poisson Distribution
from scipy.stats import poisson

mu = 3
x = np.arange(0, 10)
pmf = poisson.pmf(x, mu)

plt.bar(x, pmf)
plt.title("Poisson Distribution (=3)")
plt.xlabel("Events")
plt.ylabel("Probability")
plt.show()
```

```python
# 10. Central Limit Theorem (CLT)
# Non-normal distribution (Exponential)
data_samples = []
for _ in range(1000):
    sample = np.random.exponential(scale=2, size=30)
    sample_mean = np.mean(sample)
    data_samples.append(sample_mean)

plt.hist(data_samples, bins=30, density=True, edgecolor='black')
plt.title("Central Limit Theorem (Means of Exponential Distribution)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

```python
# Q18. Z-Score for a Given Data Point
import scipy.stats as stats

data = [50, 52, 53, 54, 55, 60, 65]
value = 54
mean = np.mean(data)
std = np.std(data)

z_score = (value - mean) / std
print("Z-score:", z_score)

# Compare to Standard Normal Distribution
prob = stats.norm.cdf(z_score)
print("Probability below this value:", prob)
```

```python

# Q19. Hypothesis Testing using Z-statistics
# One-sample Z-test
population_mean = 50
sample = np.array([52, 49, 50, 51, 48, 53, 52])
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Z-statistic: {z_stat:.3f}, p-value: {p_value:.3f}")
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
# Q20. Confidence Interval for Dataset
import scipy.stats as st

data = np.random.randint(40, 60, size=20)
mean = np.mean(data)
std_err = st.sem(data)
ci = st.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)

print("Data Mean:", mean)
print("95% Confidence Interval:", ci)
```

Q1. What is a random variable in probability theory?
A random variable is a variable whose value is determined by the outcome of a random experiment. It can take different possible values, each with an associated probability.

Q2. What are the types of random variables?
There are mainly two types: (1) Discrete Random Variable  takes a finite or countable set of values. (2) Continuous Random Variable  takes infinite possible values within an interval.

Q3. What is the difference between discrete and continuous distributions?
Discrete distribution assigns probability to specific values, while continuous distribution spreads probability over a range of values, using a density function.

Q4. What are probability distribution functions (PDF)?
A PDF describes how the probability of a continuous random variable is distributed. The area under the curve between two points gives the probability for that range.

Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
PDF shows probability density for a value range, while CDF shows cumulative probability from the start up to a certain point.

Q6. What is a discrete uniform distribution?
A distribution where all outcomes are equally likely. Example: Rolling a fair die (each side has 1/6 probability).

Q7. What are the key properties of a Bernoulli distribution?
Two outcomes: success (1) with probability p, and failure (0) with probability 1p. Mean = p, Variance = p(1p).

Q8. What is the binomial distribution, and how is it used in probability?
It models the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success.

Q9. What is the Poisson distribution and where is it applied?
It models the number of events in a fixed interval of time or space, when events happen independently and at a constant average rate.

Q10. What is a continuous uniform distribution?
A distribution where all values in a continuous range are equally likely.

Q11. What are the characteristics of a normal distribution?

Q12. What is the standard normal distribution, and why is it important?
It is a normal distribution with mean 0 and standard deviation 1, widely used for calculating probabilities and Z-scores.

Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
It states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of population distribution.

Q14. How does the Central Limit Theorem relate to the normal distribution?
It explains why averages of samples tend to follow a normal distribution even if the original data is not normal.

Q15. What is the application of Z statistics in hypothesis testing?
Z-statistics help decide whether to reject a null hypothesis when the population variance is known.

Q17. What are point estimates and interval estimates in statistics?
Point estimate gives a single best guess for a parameter, while interval estimate provides a range within which the parameter likely lies.

Q18. What is the significance of confidence intervals in statistical analysis?
They give a range of possible values for a parameter with a certain confidence level, indicating estimation accuracy.

Q19. What is the relationship between a Z-score and a confidence interval?
Z-scores define the boundaries of a confidence interval in a normal distribution.

Q20. How are Z-scores used to compare different distributions?

Q21. What are the assumptions for applying the Central Limit Theorem?
Independent samples, large enough sample size, and population with finite variance.

Q22. What is the concept of expected value in a probability distribution?
It is the long-run average value of a random variable, weighted by probabilities.
